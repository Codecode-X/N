```python
import torch
import clip
from PIL import Image

device = "cpu" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/16", device=device)
```

```python
image_path = "exp/input_images/下半身（上红下白）.png"
choices = ["Red on the bottom", "Red on the up"]
image = preprocess(Image.open(image_path)).unsqueeze(0).to(device)
text = clip.tokenize(choices).to(device)

with torch.no_grad():
    image_features = model.encode_image(image)
    text_features = model.encode_text(text)
    
    logits_per_image, logits_per_text = model(image, text)
    probs = logits_per_image.softmax(dim=-1).cpu().numpy()

print(choices)
print("Label probs:", probs) 


choices = ["bottom", "up"]
image = preprocess(Image.open(image_path)).unsqueeze(0).to(device)
text = clip.tokenize(choices).to(device)

with torch.no_grad():
    image_features = model.encode_image(image)
    text_features = model.encode_text(text)
    
    logits_per_image, logits_per_text = model(image, text)
    probs = logits_per_image.softmax(dim=-1).cpu().numpy()

print(choices)
print("Label probs:", probs) 


choices = ["white on the bottom", "white on the up"]
image = preprocess(Image.open(image_path)).unsqueeze(0).to(device)
text = clip.tokenize(choices).to(device)

with torch.no_grad():
    image_features = model.encode_image(image)
    text_features = model.encode_text(text)
    
    logits_per_image, logits_per_text = model(image, text)
    probs = logits_per_image.softmax(dim=-1).cpu().numpy()

print(choices)
print("Label probs:", probs) 

### clip主要还是关注这个图片是某个人的下半身，而没有关注颜色
```

#### 目标检测+clip

```python
"""目标检测+clip"""

from ultralytics import YOLO
import clip
import torch
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# 加载 YOLO 目标检测模型
yolo_model = YOLO("yolov8n.pt")  # 轻量级模型
image_path = "exp/input_images/raw.png"

# 目标检测
image = Image.open(image_path)
results = yolo_model(image) # 返回batchsize个result

# 遍历检测到的目标
fig, axs = plt.subplots(1, len(results[0].boxes), figsize=(15, 5))  # 创建子图

for idx, box in enumerate(results[0].boxes):
    x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())  # 获取 bbox 坐标
    cropped = image.crop((x1, y1, x2, y2))  # 裁剪人物

    # 计算位置（左、中、右）
    width = image.width
    if x1 < width / 3:
        position = "left"
    elif x2 > 2 * width / 3:
        position = "right"
    else:
        position = "middle"

    # CLIP 识别颜色
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model, preprocess = clip.load("ViT-B/32", device=device)
    
    image_tensor = preprocess(cropped).unsqueeze(0).to(device)
    text = clip.tokenize(["red", "white", "blue"]).to(device)

    with torch.no_grad():
        logits_per_image, _ = model(image_tensor, text)
        probs = logits_per_image.softmax(dim=-1).cpu().numpy()

    color = ["red", "white", "blue"][probs.argmax()]
    
    # 显示裁剪后的人物
    axs[idx].imshow(cropped)
    axs[idx].set_title(f"{position}: {results[0].names[int(box.cls)]}-{color}", fontsize=12)
    axs[idx].axis("off")

plt.tight_layout()
plt.show()
```
